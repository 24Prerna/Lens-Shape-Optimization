# Gradient Descent to Optimize Lens Curvature

* >  Optimizes the shape of a lens using multivariable calculus and numerical optimization techniques. 
* > The goal is to adjust the curvature of the lens so that it focuses light at a desired focal point.

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

* numpy (np): Used to create and manipulate arrays and matrices.
* matplotlib.pyplot: Used for visualizing the optimized lens shape.
* scipy.optimize.minimize: Used to perform numerical optimization.

## Define the Lens Surface Function

```python
def lens_surface(x, y, a, b, c):
    return a*x**2 + b*y**2 + c
```

* This function represents a parabolic lens surface in 3D space.
* It takes in x and y coordinates along with parameters a, b, and c that define the shape.
* The equation z= ax^2 + by^2 + c models how light bends based on curvature.

## Define the Error Function (Optimization Target)

```python
def error_function(params):
    a, b, c = params
    error = np.sum((lens_surface(x, y, a, b, c) - ideal_focus)**2)
    return error
```

* Optimization Goal: Minimize the difference between the actual lens surface and the ideal focal point.
* params contains (a, b, c), which we will optimize.
* The function calculates the sum of squared errors, ensuring that the lens surface best focuses light to the desired point.

 ## Generate Sample Data (Lens Surface Points)

```python
x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
ideal_focus = 10  # Ideal focal point
```

* We create a grid of points (x, y) to represent the lens surface.
* The ideal_focus is set to 10, meaning we want light to be focused at z = 10

## Optimize the Lens Parameters using Gradient-Based Method

```python
initial_guess = [0.1, 0.1, 1]
result = minimize(error_function, initial_guess, method='BFGS')
```

* initial_guess = (0.1, 0.1, 1): Starts with arbitrary values for a, b, and c.
* minimize(): Uses the BFGS (Broyden–Fletcher–Goldfarb–Shanno) algorithm, a gradient-based method to find the optimal lens shape.
* The algorithm adjusts a, b, and c to minimize the error function.

## Extract Optimized Parameters

```python
optimized_a, optimized_b, optimized_c = result.x
```

After optimization, we retrieve the best values of a, b, and c for the lens shape.

## Plot the Optimized Lens Surface

```python
z = lens_surface(x, y, optimized_a, optimized_b, optimized_c)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot optimized lens surface
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels for better understanding
ax.set_xlabel("X-axis (Lens Width)")
ax.set_ylabel("Y-axis (Lens Height)")
ax.set_zlabel("Z-axis (Curvature Depth)")
ax.set_title("Optimized Lens Surface")

# Show the plot
plt.show()
```
